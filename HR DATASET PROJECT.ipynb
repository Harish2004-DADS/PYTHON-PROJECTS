> #### 7.CALCULATE AND IDENTIFY **OUTLLIERS IN A DATAFRAME.**

- THE OUTLIERS ARE CALCULATE FOR **DETECT UNUSUALLY HIGH OR LOW VALUES IN NUMERICE COLUMNS.**
- TO **SEPARATE THE NUMERIC COLUMNS FOR ANALYZING OUTLIERS.**

```python
num_column_name=HR_DATA.select_dtypes(include=['number']).columns.tolist()
print(num_column_name)
```

- IN OUR DATASET, THERE ARE **23 NUMERICAL COLUMNS.**

```python
num_column=HR_DATA[['Age', 'Daily Rate', 'Distance From Home', 'Education', 'Environment Satisfaction', 'Hourly Rate', 'Job Involvement', 'Job Level', 'Job Satisfaction', 'Monthly Income', 'Monthly Rate', 'Num Companies Worked', 'Percent Salary Hike', 'Performance Rating', 'Relationship Satisfaction', 'Stock Option Level', 'Total Working Years', 'Training Times Last Year', 'Work Life Balance', 'Years At Company', 'Years In Current Role', 'Years Since Last Promotion', 'Years With Curr Manager']]
len(num_column.columns)
```

-USING **BOXPLOTS**, WE CAN IDENTIFY **OUTLIERS** FOR THESE **23 NUMERICAL COLUMNS AS WELL.**

```python
plt.figure(figsize=(20,30))
for i in range(len(num_column.columns)):
    plt.subplot(8,3,i+1)
    sns.boxplot(y=num_column.columns[i],data=HR_DATA)
plt.show()
```

- THE FOLLOWING PROGRAM SEPARATES THE DATASET'S COLUMNS INTO TWO CATEGORIES-**WITH OUTLIERS** AND **WITHOUT OUTLIERS**

```python
cols_with_outliers = []
cols_without_outliers = []

for col in num_column_name:
    Q1 = HR_DATA[col].quantile(0.25)
    Q3 = HR_DATA[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    if ((HR_DATA[col] < lower) | (HR_DATA[col] > upper)).any():
        cols_with_outliers.append(col)
    else:
        cols_without_outliers.append(col)

print("With outliers:", cols_with_outliers)
print(len(cols_with_outliers))
print('\n')
print("Without outliers:", cols_without_outliers)
print(len(cols_without_outliers))
```

- THIS HAVE ONLY **WITHOUT OUTLIERS COLUMNS.**

```python
cols_without_outliers=HR_DATA[['Age', 'Daily Rate', 'Distance From Home', 'Education', 'Environment Satisfaction', 'Hourly Rate', 'Job Involvement', 'Job Level', 'Job Satisfaction', 'Monthly Rate', 'Percent Salary Hike', 'Relationship Satisfaction', 'Work Life Balance']]
len(cols_without_outliers.columns)
```

- TO CALCULATE THE **INTERQUARTILE RANGE (IQR), LOWER BOUND, UPPER BOUND, AND MEDIAN** TO REPLACE THE OUTLIERS FOR THESE **10 NUMERICAL COLUMNS AS WELL.**

```python
for i in cols_with_outliers.columns:
    Q1 = HR_DATA[i].quantile(0.25)
    Q3 = HR_DATA[i].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    median = HR_DATA[i].median()
    HR_DATA[i] = HR_DATA[i].apply(lambda x: median if x < lower or x > upper else x)
```

4.HYPOTHESIS TESTING
------

- #### HYPOTHESIS TESTING TO HOUSING DATASET ANALYSIS WILL GIVE PROJECT MORE STATISTICAL DEPTH AND ANALYTICAL SKILLS.


#### **1.T-TESTING**

**1.ONE-SAMPLE T-TEST FOR MONTHLY INCOME**

```python
from scipy import stats
# Is average Monthly Income significantly different from ₹50,000?
income = HR_DATA['Monthly Income']
mean_income = np.mean(income)
print(f'THE MEAN OF MONTHLY INCOME:{mean_income}.')

np.random.seed(11)
sample_size = 10
sample_income = np.random.choice(income,sample_size)
print(f'THE SAMPLE MONTHLY INCOME:{sample_income}.')

_,p_value = stats.ttest_1samp(a=sample_income,popmean=mean_income)
print(f'THE P_VALUE FOR MONTHLY INCOME:{p_value}.')

if p_value < 0.05:
    print("Reject H₀: Renovation has a significant impact on monthly income.")
else:
    print("Fail to Reject H₀: No significant impact found.")
```

**^ INSIGHTS:**
> - Null Hypothesis (H₀): The average employee income is equal to ₹50,000.

> - Alternative Hypothesis (H₁): The average employee income is not equal to ₹50,000.

> - Result: Since p < 0.05, we reject H₀ → Income is significantly different from ₹50,000.

**2.INDEPENDENT SAMPLES T-TEST FOR GENDER VS JOB SATISFACTION**

```python
import numpy as np
from scipy.stats import ttest_ind

# Define the two groups for Glucose levels
group1 = HR_DATA[HR_DATA['Gender'] == 'Male']['Job Satisfaction']  
group2 = HR_DATA[HR_DATA['Gender'] == 'Female']['Job Satisfaction']

# Calculate and display variances
var_group1 = np.var(group1, ddof=1)  # Using ddof=1 for sample variance
var_group2 = np.var(group2, ddof=1)  # Using ddof=1 for sample variance
print("Variance of Job Satisfaction for non-diabetic group:", var_group1)
print("Variance of Job Satisfaction for diabetic group:", var_group2)

# Check if variances are approximately equal
alpha = 0.05
if abs(var_group1 - var_group2) / max(var_group1, var_group2) < 0.1:  # Allowing a 10% tolerance
    print("Variances are approximately equal; setting equal_var=True.")
    equal_var = True
else:
    print("Variances are not approximately equal; setting equal_var=False.")
    equal_var = False

# Perform the two-sample t-test with the appropriate equal_var setting
t_stat, p_val = ttest_ind(group1, group2, equal_var=equal_var)
print("t-statistic:", t_stat)
print("p-value:", p_val)
if p_val < alpha:
    print("Since p-value < alpha, we reject the null hypothesis: there is a significant difference between the groups.")
else:
    print("Since p-value >= alpha, we fail to reject the null hypothesis: there is no significant difference between the groups.")
```

**^ INSIGHTS:**
> - Null Hypothesis (H₀): There is no significant difference in job satisfaction between male and female employees.

> - Alternative Hypothesis (H₁): There is a significant difference in job satisfaction between male and female employees.

> - Result: Since p > 0.05, we fail to reject H₀ → No significant difference between genders.

**3.PAIRED SAMPLES T-TEST FOR JOB SATISFACTION AND ENVIRONMENT SATISFACTION**

```python
import numpy as np
from scipy import stats

# Example BMI values before and after treatment
job = HR_DATA['Job Satisfaction']
env = HR_DATA['Environment Satisfaction']

# Calculate the mean difference and perform paired sample t-test
mean_diff = np.mean(job - env)
t_stat, p_value = stats.ttest_1samp(job - env, 0)

print(f"Mean Difference: {mean_diff}")
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")

# Set significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Job Satisfaction and Environment Satisfaction are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two.")
```

**^ INSIGHTS:**
> - Null Hypothesis (H₀): There is no significant difference between job satisfaction and environment satisfaction.

> - Alternative Hypothesis (H₁): There is a significant difference between job satisfaction and environment satisfaction.

> - Result: Since p < 0.05, we reject H₀ → The two satisfaction levels are different.

#### **2.Z-TESTING**

**1.ONE-SAMPLE Z-TEST FOR AGE**

```python
import numpy as np
from scipy.stats import norm
from statsmodels.stats.weightstats import ztest
HR_DATA1 = HR_DATA.dropna(subset=['Age']).sample(100, replace=True)

pop_mean = 35

x = HR_DATA1['Age'].mean()
print("Sample Mean:", x)

se = HR_DATA1['Age'].std()
print("Sample Standard Deviation:", se)
zstat, p_val = ztest(HR_DATA1['Age'], value=pop_mean)
print("z-statistic:", zstat)
print("p-value:", p_val)

alpha = 0.05
if p_val < alpha:

    print("Reject H₀: Average age is significantly different from 35.")
else:

    print("Fail to reject H₀: No significant difference in average age.")
```

**^ INSIGHTS:**
> - Null Hypothesis (H₀): The average age of employees is equal to 35 years.

> - Alternative Hypothesis (H₁): The average age of employees is significantly different from 35 years.

> - Result: The test shows no significant difference in average age compared to 35 (Fail to reject H₀).

**2.TWO-SAMPLE Z-TEST FOR GENDER VS DISTANCE FROM HOME**

```python
diabetes_positive = HR_DATA[HR_DATA["Gender"] == "Male"]["Distance From Home"]
diabetes_negative = HR_DATA[HR_DATA["Gender"] == "Female"]["Distance From Home"]

positive_mean = diabetes_positive.mean()
negative_mean = diabetes_negative.mean()
positive_std = diabetes_positive.std(ddof=1)
negative_std = diabetes_negative.std(ddof=1)

# Sample sizes
positive_n = len(diabetes_positive)
negative_n = len(diabetes_negative)

# Z = (x̄1 - x̄2) / √((s1^2 / n1) + (s2^2 / n2))
z = (positive_mean - negative_mean) / np.sqrt((positive_std**2 / positive_n) + (negative_std**2 / negative_n))

# Calculate p-value for two-tailed test
p = 2 * (1 - stats.norm.cdf(abs(z)))

print("Mean Glucose (Diabetes Positive):", positive_mean)
print("Mean Glucose (Diabetes Negative):", negative_mean)
print("Z-score:", z)
print("p-value:", p)

alpha = 0.05
if p < alpha:
    print("Reject H₀: Distance from home differs between genders.")
else:
    print("Fail to reject H₀: No significant difference in distance from home.")
```

**^ INSIGHTS:**
> - Null Hypothesis (H₀): There is no significant difference in distance from home between male and female employees.

> - Alternative Hypothesis (H₁): There is a significant difference in distance from home between male and female employees.

> - Result: The test shows no significant difference in distance from home between genders (Fail to reject H₀).

**3.PAIRED Z-TEST FOR YEARS AT COMPANY AND YEARS WITH CURRENT MANAGER**

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

col_before =HR_DATA['Years At Company']
col_after = HR_DATA['Years With Curr Manager']

HR_DATA['Difference'] = col_after - col_before

mean_diff = HR_DATA['Difference'].mean()
std_diff = HR_DATA['Difference'].std()
n = len(HR_DATA)
sigma_d = 5 

z_score = mean_diff / (sigma_d / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Paired Z-Test between '{col_before}' and '{col_after}'")
print(f"Mean of Differences: {mean_diff:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Significant difference between the two columns.")
else:
    print("Fail to reject null hypothesis: No significant difference between the two columns.")
```

**^ INSIGHTS:**
> - Null Hypothesis (H₀): There is no significant difference between years at the company and years with the current manager.

> - Alternative Hypothesis (H₁): There is a significant difference between years at the company and years with the current manager.

> - Result: The test shows a significant difference between years at the company and years with the current manager (Reject H₀).

#### **3.ANOVA TESTING**

**1.ONE-WAY ANOVA FOR JOB ROLE AND YEARS AT COMPANY**

```python
import pandas as pd
import scipy.stats as stats

# Drop missing values
df_col =HR_DATA[['Job Role', 'Years At Company']]

# Group data by Job Role
groups = [group['Years At Company'].values for name, group in df_col.groupby('Job Role')]

# One-Way ANOVA
f_stat, p_val = stats.f_oneway(*groups)
alpha = 0.05

print(f"F-statistic: {f_stat:.3f}, P-value: {p_val:.3f}")
if p_val < alpha:
    print("Conclusion: We reject the null hypothesis.")
    print("Reject H₀: Job Role has a significant effect on Years at Company.")
else:
    print("Conclusion: We fail to reject the null hypothesis.")
    print("Fail to reject H₀: No significant difference across Job Roles.")
```

**^ INSIGHTS:**
> - Null Hypothesis (H₀): Job role has no significant effect on years at the company.

> - Alternative Hypothesis (H₁): Job role has a significant effect on years at the company.

> - Result: The test shows a significant difference (Reject H₀).

**2.TWO-WAY ANOVA FRO PERFORMANCE RATING DEPARTMENT AND EDUCATION FIELD**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Drop missing values
df_col = HR_DATA[['Performance Rating', 'Department', 'Education Field']]

# Build model
model = ols('Q("Performance Rating") ~ C(Department) + C(Q("Education Field")) + C(Department):C(Q("Education Field"))', data=df_col).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)

# Interpretation
alpha = 0.05
for factor in ['C(Department)', 'C(Q("Education Field"))', 'C(Department):C(Q("Education Field"))']:
    p_val = anova_table.loc[factor, 'PR(>F)']
    if p_val < alpha:
        print(f"Reject H₀: {factor} has a significant effect on Performance Rating.")
    else:
        print(f"Fail to reject H₀: {factor} has no significant effect.")
```

**^ INSIGHTS:**
> - Null Hypothesis (H₀): Department and Education Field do not affect performance rating.

> - Alternative Hypothesis (H₁): Department or Education Field affects performance rating.

> - Result: Education Field and the interaction of Department × Education Field have a significant effect, but Department alone does not.
